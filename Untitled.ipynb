```python
i = 0
for doc in cursor:
    d = snappy.decompress(doc['data'])
    w = np.fromstring(d, dtype='<i2')
    plt.plot(np.arange(len(w))  * 2, w)
    plt.title("Waveform from module %d:%d, time %d-%d" % (
            doc['module'], doc['channel'], doc['time'], doc['endtime']))
    plt.show()
    
    i += 1
    print(doc['channel'])
    if i > 10:
        break
```
