```python
from pymongo import MongoClient
import matplotlib.pyplot as plt
```

```python
client = MongoClient()
```

```python
client.database_names()
```

```python
client['xamsdata0'].collection_names()
```

```python
dbname = 'xamsdata0'
collname = 'gamma_test'
```

```python
cursor = client[dbname][collname].find()
```

```python
[{'pmt_position': ch, 
         'position': {'x':0, 'y':0},
         'array': {0: 'top', 3:'bottom'}.get(ch, 'other')} for ch in range(8)]
```

```python
import snappy
import numpy as np
%matplotlib inline


```

```python
i = 0
for doc in cursor:
    d = snappy.decompress(doc['data'])
    w = np.fromstring(d, dtype='<i2')
    if 1 <= i <= 20: 
        plt.plot(np.arange(len(w))  * 2, w)
        plt.title("Waveform from module %d:%d, time %d-%d" % (
                doc['module'], doc['channel'], doc['time'], doc['endtime']))
        plt.xlim(2800,4000)
        plt.show()
    
    i += 1
```
