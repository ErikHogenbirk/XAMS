```python
filename = '/home/xams/xams/darkcounts/900vt8180hf30.hdf5'

import os
if not os.path.exists(filename):
    raise FileNotFoundError("not found")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
store = pd.HDFStore(filename)
data= store['hits']
```

```python
plt.scatter(data['area'], data['height'])
plt.xscale('log')
plt.yscale('log')
plt.xlabel("Area (electrons)")
plt.ylabel("Height (electrons)")
plt.grid(alpha=0.2, linestyle='-')
plt.xlim([1000000, 7000000])
```

```python
plt.hist(data['area'], bins=np.logspace(5, 9, 100));
plt.xscale('log')
```

```python
jigg = np.transpose([range(int(len(data['area']))), data['area'], data['height']])
```

```python
result = [int(t[0]) for t in jigg if t[1]>4000000 and t[1]<6000000 and t[2]>2000000 and t[2]<4000000]
result
```
