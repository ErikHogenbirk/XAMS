```python
# settings
n_events = 1000
tts      = 10
source = 'wimp'
wimp_mass = 20

```

```python
# Derived properties
if source == 'er':
    band = 'er'
    base_name = 'er_tts%.1fns_%dk' % (tts, n_events / 1000)
elif source == 'wimp':
    band = 'nr'
    base_name = 'wimp_%dGeV_tts%.1fns_%dk' % (wimp_mass, tts, n_events / 1000)
else:
    raise NotImplementedError('Only sources can be er or wimp...')
```

```python
print('Working on base name %s' % base_name)
```

```python
base_model.config['wimp_mass'] = wimp_mass
model = bi.Model(base_model.config)
if source == 'er':
    generator = model.sources[0]
elif source == 'wimp':
    generator = model.sources[-1] 
```

```python
# df data
df = pd.DataFrame(generator.simulate(n_events))
add_props(df, band)
```
