```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
t = np.linspace(0, 639*2, 640) - 340 * 2
# Aligned on sample 340
```

```python
dset_ch0 = '170323_103732' # Top PMT
dset_ch1 = '170323_104831' # Bottom PMT
```

```python
pulse_ch0 = pickle.load(open('%s_ch%d.pickle' % (dset_ch0, 0) , 'rb'))
pulse_ch1 = pickle.load(open('%s_ch%d.pickle' % (dset_ch1, 1) , 'rb'))
```

```python
for channel, pulse, desc in zip([0, 1], [pulse_ch0, pulse_ch1], ['top PMT', 'bottom PMT']):
    plt.plot(t, pulse, marker = '.', markersize = 5)
    plt.xlim(-100, 400)
    plt.axvline(0, ls='--', color='gray', lw = 1)
    plt.xlabel('Time (ns)')
    plt.title('Average pulse channel %d (%s)' % (channel, desc))
    plt.ylabel('Normalized amplitude')
    plt.show()
```
