```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dset_ch0 = '170323_103732' # Top PMT
dset_ch1 = '170323_104831' # Bottom PMT
```

```python
def gaus_trunc(x, mu, sigma):
    return (x > 0) * np.exp( - (x - mu)**2 / (2 * sigma**2))
```

```python
norm0, popt0, perr0 = pickle.load(open(dset_ch0 + '_ch0_function.pickle', 'rb'))
norm1, popt1, perr1 = pickle.load(open(dset_ch1 + '_ch1_function.pickle', 'rb'))
```

```python
x_plot = np.linspace(-1, 4, 100)
for norm, popt, channel in zip((norm0, norm1), (popt0, popt1), (0, 1)):
    plt.plot(x_plot, norm * gaus_trunc(x_plot, *popt), label = 'Channel %d' % channel)
plt.legend()
plt.ylabel('Differential area or something fancy like that')
plt.xlabel('Area (p.e.)')
# Fit range used, do not put a lot of confidence in distribution below the red line
for _l in (0.6, 2):
    plt.axvline(_l, lw= 1, ls='--', color='red')
plt.show()
```
