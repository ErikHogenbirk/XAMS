### Extra properties

```python
d['time_since_start'] = (d['event_time'] - d['event_time'].iloc[0]) * 1e-9
d['s2_bottom'] = d['s2'] * (1 - d['s2_area_fraction_top'])

def ly(dt):
    return 240 + 1*dt + 0.035*dt**2
avg_s1 = np.average([ly(dt) for dt in np.linspace(0, 59.5, 100)])
def s1_correction(dt_ns, avg_s1):
    dt = dt_ns*0.001
    return avg_s1 / ly(dt)    

d['cs1'] = d['s1'] * s1_correction(d['drift_time'], avg_s1)
d['cs1_low'] = d['s1'] * s1_correction(d['drift_time'] * 73/59, avg_s1)
```
