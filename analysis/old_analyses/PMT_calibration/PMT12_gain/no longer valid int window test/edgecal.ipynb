```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Standard right-left edges are 650-750ns

rightedge = [820, 780, 750, 730, 720, 710, 700]
gain1r = [5.73, 5.72, 5.61, 5.56, 5.57, 5.21, 3.75]
gain2r = [4.02, 3.93, 3.97, 4.00, 3.65, 2.70, 1.87]
leftedge = [550, 600, 650, 680, 690, 700]
gain1l = [5.75, 5.70, 5.61, 5.65, 5.31, 2.87]
gain2l = [3.90, 3.92, 3.97, 4.01, 3.57, 2.3]
```

```python
plt.plot(rightedge, gain1r, color = 'red', marker = '.')
plt.plot(rightedge, gain2r, color = 'blue', marker = '.')
plt.xlabel('right edge')
plt.ylabel('gain')
plt.savefig('rightedge.eps', format='eps')
plt.show()
plt.plot(leftedge, gain1l, color = 'red', marker = '.')
plt.plot(leftedge, gain2l, color = 'blue', marker = '.')
plt.xlabel('left edge')
plt.ylabel('gain')
plt.savefig('leftedge.eps', format='eps')
plt.show()
```
