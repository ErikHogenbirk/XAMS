```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Standard right-left edges are 650-750ns

rightedge = [690, 700, 720, 750, 770, 800, 850]
gain1r = [3.90, 5.00, 5.28, 5.34, 5.59, 5.58, 5.60]
gain2r = [1.81, 2.69, 3.74, 3.67, 3.58, 3.69, 3.78]
leftedge = [550, 600, 650, 670, 680, 690, 700]
gain1l = [5.53, 5.58, 5.53, 5.60, 4.77, 2.05, 0]
gain2l = [3.72, 3.69, 3.67, 3.72, 3.05, 1.81, 0]
```

```python
plt.plot(rightedge, gain1r, color = 'red', marker = '.')
plt.plot(rightedge, gain2r, color = 'blue', marker = '.')
plt.xlabel('right edge')
plt.ylabel('gain')
plt.savefig('rightedge.eps', format='eps')
plt.show()
plt.plot(leftedge, gain1l, color = 'red', marker = '.')
plt.plot(leftedge, gain2l, color = 'blue', marker = '.')
plt.xlabel('left edge')
plt.ylabel('gain')
plt.savefig('leftedge.eps', format='eps')
plt.show()
```
