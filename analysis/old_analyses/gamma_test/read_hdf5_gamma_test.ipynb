```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from scipy.optimize import curve_fit
%matplotlib inline
import os
import math
```

```python
filename1 = '/home/xams/xams/gamma_test/5_min_background.hdf5'
filename2 = '/home/xams/xams/gamma_test/5_min_1_sheet.hdf5'
filename3 = '/home/xams/xams/gamma_test/5_min_6mm.hdf5'
    
store1 = pd.HDFStore(filename1)
store2 = pd.HDFStore(filename2)
store3 = pd.HDFStore(filename3)
data1= store1['all_hits']
data2= store2['all_hits']
data3= store3['all_hits']

print(len(data1))
print(len(data2))
print(len(data3))
```

```python
histboundaries = [0, 3]
histbinwidth = 0.02
histbins = np.arange(histboundaries[0], histboundaries[1]+histbinwidth, histbinwidth)
centers = np.arange(histboundaries[0]+histbinwidth/2.0, histboundaries[1], histbinwidth)

melecs1 = data1['area']/1e6*1.332/895
melecs2 = data2['area']/1e6*1.332/895
melecs3 = data3['area']/1e6*1.332/895

hist = plt.hist([melecs1, melecs2, melecs3], bins=histbins, histtype='step')

plt.xlabel("Energy (MeV)")
plt.ylabel("counts")
plt.suptitle("Background gammas with 0,3,6mm lead (74k, 15k, 10k)")
plt.yscale('log')
# plt.xlim(850,950)
plt.ylim(0,10000)
plt.savefig('bgreduc.eps', format='eps')
plt.show()
```
